```python
import pytesseract, cv2, numpy as np
from pdf2image import convert_from_path
from PIL import Image

def preprocess_image(img):
    """Enhances image for best OCR accuracy (grayscale, thresholding, denoise)."""
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.medianBlur(gray, 3)
    thresh = cv2.adaptiveThreshold(
        blur, 255,
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 45, 11
    )
    return thresh

def ocr_pdf_or_image(file_path, out_folder="data/ocr_text"):
    os.makedirs(out_folder, exist_ok=True)
    texts = []
    if file_path.lower().endswith(".pdf"):
        pages = convert_from_path(file_path, dpi=300)
        for idx, page in enumerate(pages):
            pg_img_path = os.path.join(out_folder, f"{os.path.basename(file_path)}_page{idx+1}.png")
            page.save(pg_img_path, "PNG")
            img = cv2.imread(pg_img_path)
            proc_img = preprocess_image(img)
            text = pytesseract.image_to_string(proc_img, lang="eng+hin")
            texts.append(text)
            # Save OCR text with provenance
            with open(f"{pg_img_path}.txt", "w", encoding="utf-8") as f:
                f.write(text)
    else:
        img = cv2.imread(file_path)
        proc_img = preprocess_image(img)
        text = pytesseract.image_to_string(proc_img, lang="eng+hin")
        texts.append(text)
        with open(f"{file_path}.txt", "w", encoding="utf-8") as f:
            f.write(text)
    # Collate all pages if PDF
    fulltext = "\n".join(texts)
    print(f"OCR complete: {len(texts)} page(s)")
    return fulltext
```

```python
import os
import glob
import re
import pandas as pd

ocr_out_folder = "data/ocr_text"
registry_path = "data/document_registry.csv"
output_csv = "data/chunked_cleaned_text.csv"

# Helper: get document base name up to _page
def document_basename(filename):
    return filename.rsplit("_page", 1)[0]

# Step 1: Gather and process all OCR .txt files
all_chunks = []
txt_files = sorted(glob.glob(os.path.join(ocr_out_folder, "*.txt")))

for txt_file in txt_files:
    basename = os.path.basename(txt_file)
    doc_base = document_basename(basename)
    page_match = re.search(r"_page(\d+)", basename)
    page_num = int(page_match.group(1)) if page_match else None

    with open(txt_file, encoding="utf-8") as f:
        text = f.read().strip()

    # Step 2: Paragraph split (refine if needed)
    # Split on two or more newlines, keep only non-empty paragraphs
    paragraphs = [p.strip() for p in re.split(r'\n{2,}', text) if p.strip()]
    for para_idx, para in enumerate(paragraphs):
        chunk_id = f"{doc_base}_p{page_num}_para{para_idx+1:03d}"
        all_chunks.append({
            "chunk_id": chunk_id,
            "doc_base": doc_base,
            "page_num": page_num,
            "para_idx": para_idx+1,
            "text": para,
            "ocr_file": basename,
        })

# Step 3: Attach doc_id from registry using exact doc_base
if os.path.exists(registry_path):
    registry = pd.read_csv(registry_path)
    reg_map = dict(zip(registry["filename"], registry["doc_id"]))
    for chunk in all_chunks:
        fname = chunk["doc_base"]        # already has extension!
        chunk["doc_id"] = reg_map.get(fname, None)
else:
    for chunk in all_chunks:
        chunk["doc_id"] = None


# Step 4: Save all chunks + metadata as CSV (very ML/tracing friendly)
df = pd.DataFrame(all_chunks)
df.to_csv(output_csv, index=False)
print(f"Saved {len(df)} chunks in {output_csv}. Example rows:")
print(df.head(5))
```

```python
from transformers import pipeline
import pandas as pd

# Load enriched chunks with keywords
df = pd.read_csv("data/chunked_with_keywords.csv")

# Initialize summarization pipeline with a good pretrained model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Summarize each chunk text (optional: summarize grouped chunks later for sections)
def summarize_text(text):
    if not isinstance(text, str) or len(text.strip()) == 0:
        return ""
    max_input_length = 1024 * 1.3
    truncated_text = text[:int(max_input_length)]

    # Calculate input token length roughly (or character length fallback)
    length = len(truncated_text.split())

    # Set max_length proportional to input size, not exceeding 250
    max_len = max(8, min(250, length // 2))

    summary = summarizer(truncated_text, max_length=max_len, min_length=5, do_sample=False)
    return summary[0]['summary_text']


# Apply summarization (can be slow; consider batching or GPU)
df['summary'] = df['text'].apply(summarize_text)

# Save final enriched file
df.to_csv("data/chunked_with_keywords_and_summary.csv", index=False)

print(df[['chunk_id', 'summary']].head())
```

```python
import pandas as pd
import re
from transformers import AutoTokenizer

# Load your enriched data with keywords and summaries
df = pd.read_csv("data/chunked_with_keywords_and_summary.csv")

tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")

def num_tokens(text):
    return len(tokenizer.encode(str(text), truncation=False))

# 1. Normalize chunk sizes: merge short chunks and split long chunks
min_tokens = 60
max_tokens = 300
overlap = 50

merged_chunks = []
buffer = None

for idx, row in df.iterrows():
    tokens = num_tokens(row['text'])
    if buffer is not None:
        combined = buffer['text'] + " " + row['text']
        if num_tokens(combined) < min_tokens:
            buffer['text'] = combined
            buffer['chunk_id'] += "+" + row['chunk_id']
            continue
        else:
            merged_chunks.append(buffer)
            buffer = row.copy()
    else:
        buffer = row.copy()
if buffer is not None:
    merged_chunks.append(buffer)

# Now split long chunks with overlap
final_chunks = []
for row in merged_chunks:
    text = row['text']
    tokens = tokenizer.encode(text, truncation=False)
    if len(tokens) <= max_tokens:
        final_chunks.append(row)
    else:
        starts = range(0, len(tokens), max_tokens - overlap)
        for i, start in enumerate(starts):
            end = start + max_tokens
            chunk_tokens = tokens[start:end]
            chunk_text = tokenizer.decode(chunk_tokens, skip_special_tokens=True)
            new_row = row.copy()
            new_row['text'] = chunk_text
            new_row['chunk_id'] = f"{row['chunk_id']}_split{i+1:02d}"
            final_chunks.append(new_row)

df_refined = pd.DataFrame(final_chunks)

# 2. Extract heading/section if present at the start of chunk text
def extract_heading(text):
    lines = str(text).split('\n')
    for line in lines[:2]:
        line_stripped = line.strip()
        if line_stripped.isupper() and len(line_stripped.split()) < 12:
            return line_stripped
        if re.match(r'^(SECTION|CLAUSE|CHAPTER|अनुभाग|धारा|PART)[\s\.\d\w]*', line_stripped, re.I):
            return line_stripped
    return ""

df_refined['detected_heading'] = df_refined['text'].apply(extract_heading)

# 3. Detect key obligation verbs for classification
obligation_verbs = ["must", "shall", "required", "prohibited", "may not", "not allowed", "should", "cannot", "will be", "obliged"]
def detect_obligation_verbs(text):
    text_lower = str(text).lower()
    return [verb for verb in obligation_verbs if verb in text_lower]

df_refined['obligation_verbs'] = df_refined['text'].apply(detect_obligation_verbs)

# 4. Basic language detection by Unicode range (Hindi vs English)
def detect_language(text):
    # Check for Devanagari unicode block presence
    if any('\u0900' <= c <= '\u097F' for c in text):
        return "hi"
    else:
        return "en"

df_refined['language'] = df_refined['text'].apply(detect_language)

# Save the refined dataset for downstream use
df_refined.to_csv("data/chunked_refined_metadata.csv", index=False)

print(df_refined.head(10)[['chunk_id', 'detected_heading', 'obligation_verbs', 'language']])
```

```python
import re

def label_chunk(text):
    # Alert pattern
    if re.search(r"\b(alert|urgent|immediately|notice|attention|सावधान|सूचना|तत्काल)\b", text, re.I):
        return "alert"
    # Obligation pattern (strong verbs)
    elif re.search(r"\b(shall|must|required to|responsible for|should|अवश्य|अनिवार्य|करना होगा|पाबंद|not allowed|prohibited|prohibit)\b", text, re.I):
        return "obligation"
    # Regulation pattern
    elif re.search(r"\b(regulation|rule|act|order|प्रावधान|विनियमन|अधिनियम|नियम)\b", text, re.I):
        return "regulation"
    else:
        return "other"

df['chunk_type'] = df['text'].apply(label_chunk)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load data and embeddings
df = pd.read_pickle("data/chunked_with_embeddings.pkl")
embeddings = np.load("data/chunk_embeddings.npy")

# Sparse retrieval function (based on keywords)
def find_sparse_candidates(query_idx, max_candidates=30):
    query_row = df.iloc[query_idx]
    q_keywords = set(word.lower() for word in query_row['keywords'] if isinstance(word, str))
    if not q_keywords:
        return []
    candidates = []
    for idx, row in df.iterrows():
        if idx == query_idx:
            continue
        c_keywords = set(word.lower() for word in row['keywords'] if isinstance(word, str))
        if q_keywords & c_keywords:
            candidates.append(idx)
    return candidates[:max_candidates]

# Dense ranking function using cosine similarity
def dense_ranking(query_idx, candidate_idxs, top_k=5):
    q_emb = embeddings[query_idx:query_idx+1]
    cand_emb = embeddings[candidate_idxs]
    sims = cosine_similarity(q_emb, cand_emb)[0]
    sorted_idx = np.argsort(sims)[::-1]
    ranked_idxs = [candidate_idxs[i] for i in sorted_idx[:top_k]]
    ranked_scores = [sims[i] for i in sorted_idx[:top_k]]
    return ranked_idxs, ranked_scores

# Veto filtering (example: keep links with same language only)
def veto_filter(query_idx, candidate_idxs):
    q_lang = df.iloc[query_idx]['language']
    filtered = [idx for idx in candidate_idxs if df.iloc[idx]['language'] == q_lang]
    return filtered

# Build all links
all_links = []
for i in range(len(df)):
    sparse_cands = find_sparse_candidates(i)
    if not sparse_cands:
        continue
    dense_topk, scores = dense_ranking(i, sparse_cands)
    valid_links = veto_filter(i, dense_topk)
    for idx in valid_links:
        score = scores[dense_topk.index(idx)]
        all_links.append({
            "source": df.iloc[i]['chunk_id'],
            "target": df.iloc[idx]['chunk_id'],
            "score": score,
            "link_type": "semantic"
        })
    if i % 50 == 0:
        print(f"Processed {i}/{len(df)} chunks")

# Save to CSV for graph ingestion
links_df = pd.DataFrame(all_links)
links_df.to_csv("data/full_semantic_links.csv", index=False)

print(f"Total links created: {len(links_df)}")
print(links_df.head())
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

df_chunks = pd.read_pickle("data/chunked_with_embeddings.pkl")
links_df = pd.read_csv("data/full_semantic_links.csv")

# Ensure keywords column is processed as Python sets
def parse_keywords(val):
    if pd.isnull(val): return set()
    if isinstance(val, str): return set([w.strip() for w in val.split(',') if w.strip()])
    if isinstance(val, list) or isinstance(val, set): return set(val)
    return set()

df_chunks['keywords_set'] = df_chunks.get('keywords', pd.Series([""]*len(df_chunks))).apply(parse_keywords)

# Maps
chunk_to_doc = dict(zip(df_chunks['chunk_id'], df_chunks['doc_base'].apply(lambda x: x.split('/')[-1] if '/' in x else x)))
chunk_to_keywords = dict(zip(df_chunks['chunk_id'], df_chunks['keywords_set']))

# Gather info for each doc-doc edge
doc_edges = {}
for _, row in links_df.iterrows():
    doc_a = chunk_to_doc[row['source']]
    doc_b = chunk_to_doc[row['target']]
    if doc_a == doc_b:
        continue
    key = tuple(sorted([doc_a, doc_b]))
    doc_edges.setdefault(key, []).append({'score': row['score'], 'source': row['source'], 'target': row['target']})

# Build edges with best pair and common keywords
edges = []
for (doc_a, doc_b), links in doc_edges.items():
    scores = [l['score'] for l in links]
    avg_score = sum(scores) / len(scores)
    n = len(scores)
    # Most similar chunk pair
    best_link = max(links, key=lambda d: d['score'])
    src, tgt = best_link['source'], best_link['target']
    # Keywords intersection
    kw_src = chunk_to_keywords.get(src, set())
    kw_tgt = chunk_to_keywords.get(tgt, set())
    shared_kw = ', '.join(sorted(kw_src & kw_tgt)) if kw_src and kw_tgt else ''
    edges.append({
        'doc_a': doc_a,
        'doc_b': doc_b,
        'avg_score': avg_score,
        'count': n,
        'max_score': best_link['score'],
        'src_chunk': src,
        'tgt_chunk': tgt,
        'shared_kw': shared_kw
    })

# Construct Graph
G = nx.Graph()
for e in edges:
    G.add_edge(
        e['doc_a'], e['doc_b'],
        weight=e['count'],
        avg_score=e['avg_score'],
        max_score=e['max_score'],
        src_chunk=e['src_chunk'],
        tgt_chunk=e['tgt_chunk'],
        shared_kw=e['shared_kw']
    )

doc_chunk_counts = df_chunks['doc_base'].apply(lambda x: x.split('/')[-1] if '/' in x else x).value_counts()
node_sizes = [200 + 40 * doc_chunk_counts.get(doc, 1) for doc in G.nodes]
colors = plt.cm.tab10.colors
node_color_map = {doc: colors[i % len(colors)] for i, doc in enumerate(G.nodes)}

plt.figure(figsize=(12,8))
pos = nx.spring_layout(G, k=0.9)
nx.draw_networkx_nodes(G, pos, node_color=[node_color_map[doc] for doc in G.nodes], node_size=node_sizes)
nx.draw_networkx_edges(
    G, pos,
    width=[2 + 2 * G[u][v]['weight']/max(1,max(nx.get_edge_attributes(G,'weight').values())) for u,v in G.edges],
    alpha=0.7
)
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

# Edge labels: show average score, best score/chunk pair, shared keywords
edge_labels = {}
for u, v, d in G.edges(data=True):
    label = f"avg: {d['avg_score']:.2f}\nmax: {d['max_score']:.2f}\n{d['src_chunk']} → {d['tgt_chunk']}"
    if d['shared_kw']:
        label += f"\nkw: {d['shared_kw']}"
    edge_labels[(u, v)] = label

nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=7, label_pos=0.6)

plt.title("Semantic Document Network\n(Shared Keywords & Best Chunks on Edges)")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
from pyvis.network import Network
import pandas as pd

df_chunks = pd.read_pickle("data/chunked_with_embeddings.pkl")
links_df = pd.read_csv("data/full_semantic_links.csv")

# Clean up/parse keywords
import ast
def clean_keywords(raw_kw):
    if isinstance(raw_kw, list): return set([str(k).strip() for k in raw_kw if k and str(k).strip()])
    if isinstance(raw_kw, str):
        try:
            parsed = ast.literal_eval(raw_kw)
            if isinstance(parsed, (list, set)): return set([str(k).strip() for k in parsed if k and str(k).strip()])
            return set([raw_kw.strip()])
        except: return set([raw_kw.strip()])
    return set()

df_chunks['keywords_set'] = df_chunks['keywords'].apply(clean_keywords)
chunk_to_doc = dict(zip(df_chunks['chunk_id'], df_chunks['doc_base'].apply(lambda x: x.split('/')[-1] if '/' in x else x)))
chunk_to_keywords = dict(zip(df_chunks['chunk_id'], df_chunks['keywords_set']))

# Prepare edge metadata
doc_edges = {}
for _, row in links_df.iterrows():
    doc_a = chunk_to_doc[row['source']]
    doc_b = chunk_to_doc[row['target']]
    if doc_a == doc_b:
        continue
    key = tuple(sorted([doc_a, doc_b]))
    doc_edges.setdefault(key, []).append({'score': row['score'], 'source': row['source'], 'target': row['target']})

edges = []
for (doc_a, doc_b), links in doc_edges.items():
    scores = [l['score'] for l in links]
    avg_score = sum(scores) / len(scores)
    n = len(scores)
    best_link = max(links, key=lambda d: d['score'])
    src, tgt = best_link['source'], best_link['target']
    kw_src = chunk_to_keywords.get(src, set())
    kw_tgt = chunk_to_keywords.get(tgt, set())
    shared_kw = ', '.join(sorted(kw_src & kw_tgt)) if kw_src and kw_tgt else ''
    edges.append({
        'doc_a': doc_a,
        'doc_b': doc_b,
        'avg_score': avg_score,
        'count': n,
        'max_score': best_link['score'],
        'src_chunk': src,
        'tgt_chunk': tgt,
        'shared_kw': shared_kw
    })

# PyVis interactive network
net = Network(notebook=False, height="750px", width="100%", bgcolor="#222222", font_color="white")

# Nodes: size by #chunks, label doc name
doc_chunk_counts = df_chunks['doc_base'].apply(lambda x: x.split('/')[-1] if '/' in x else x).value_counts().to_dict()
for doc in set([e['doc_a'] for e in edges] + [e['doc_b'] for e in edges]):
    net.add_node(doc, label=doc, title=f"Document: {doc}<br>Chunks: {doc_chunk_counts.get(doc, 1)}", size=25+2*doc_chunk_counts.get(doc, 1))

# Edges: all metadata as hover/tooltip
for e in edges:
    rel = f"""
    <b>{e['doc_a']} ↔ {e['doc_b']}</b><br>
    <b>Avg similarity:</b> {e['avg_score']:.2f}<br>
    <b>Strongest connection:</b> {e['src_chunk']} → {e['tgt_chunk']}<br>
    <b>Max similarity:</b> {e['max_score']:.2f}<br>
    <b>Shared keywords:</b> {e['shared_kw'] or 'None'}<br>
    <b>Total connections:</b> {e['count']}
    """
    net.add_edge(
        e['doc_a'],
        e['doc_b'],
        value=e['avg_score'],        # edge thickness scales with similarity
        title=rel,                  # shown on hover
        label=f"{e['avg_score']:.2f}"
    )

# Save to HTML file
net.show("semantic_doc_network.html")
print("Your interactive semantic network HTML is ready: open semantic_doc_network.html")
```
