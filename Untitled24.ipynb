```python
import os, hashlib, magic, pandas as pd
from datetime import datetime

REGISTRY_PATH = "data/document_registry.csv"
if os.path.exists(REGISTRY_PATH):
    registry = pd.read_csv(REGISTRY_PATH)
else:
    registry = pd.DataFrame(columns=[
        "doc_id", "source_url", "filename", "file_hash", "file_type", "date_added", "version_date"
    ])

def compute_file_hash(file_path):
    h = hashlib.sha256()
    with open(file_path, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            h.update(chunk)
    return h.hexdigest()

def add_to_registry(file_path, source_url=None, version_date=None):
    global registry
    file_hash = compute_file_hash(file_path)
    file_type = magic.from_file(file_path, mime=True)
    doc_id = f"DOC-{len(registry)+1:04d}"
    registry = pd.concat([registry, pd.DataFrame([{
        "doc_id": doc_id,
        "source_url": source_url,
        "filename": os.path.basename(file_path),
        "file_hash": file_hash,
        "file_type": file_type,
        "date_added": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        "version_date": version_date
    }])], ignore_index=True)
    registry.to_csv(REGISTRY_PATH, index=False)
    print(f"[OK] Added: {os.path.basename(file_path)} → ID {doc_id}")

for fname in os.listdir(REG_AI_PATH):
    src = os.path.join(REG_AI_PATH, fname)
    dst = os.path.join("data/raw", fname)
    if os.path.isfile(src):
        os.system(f'cp "{src}" "{dst}"')
        add_to_registry(dst)
```

```python
import pytesseract, cv2
from pdf2image import convert_from_path
import glob

def preprocess_image(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.medianBlur(gray, 3)
    thresh = cv2.adaptiveThreshold(
        blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 45, 11)
    return thresh

def ocr_pdf_or_image(file_path, out_folder="data/ocr_text"):
    import os
    os.makedirs(out_folder, exist_ok=True)
    texts = []
    if file_path.lower().endswith(".pdf"):
        pages = convert_from_path(file_path, dpi=300)
        for idx, page in enumerate(pages):
            pg_img_path = os.path.join(out_folder, f"{os.path.basename(file_path)}_page{idx+1}.png")
            page.save(pg_img_path, "PNG")
            img = cv2.imread(pg_img_path)
            proc_img = preprocess_image(img)
            text = pytesseract.image_to_string(proc_img, lang="eng+hin")
            texts.append(text)
            with open(f"{pg_img_path}.txt", "w", encoding="utf-8") as f:
                f.write(text)
    else:
        img = cv2.imread(file_path)
        proc_img = preprocess_image(img)
        text = pytesseract.image_to_string(proc_img, lang="eng+hin")
        texts.append(text)
        with open(f"{file_path}.txt", "w", encoding="utf-8") as f:
            f.write(text)
    return "\n".join(texts)

# OCR all files
ocr_out_folder = "data/ocr_text"
os.makedirs(ocr_out_folder, exist_ok=True)
for file_path in glob.glob("data/raw/*"):
    print(file_path)
    ocr_pdf_or_image(file_path, out_folder=ocr_out_folder)
```

```python
import pandas as pd
import re
import glob

all_chunks = []
def document_basename(filename):
    return filename.rsplit("_page", 1)[0]

txt_files = sorted(glob.glob("data/ocr_text/*.txt"))
for txt_file in txt_files:
    basename = os.path.basename(txt_file)
    doc_base = document_basename(basename)
    page_match = re.search(r"_page(\d+)", basename)
    page_num = int(page_match.group(1)) if page_match else None
    with open(txt_file, encoding="utf-8") as f:
        text = f.read().strip()
    # Split on 2+ newlines for paragraphs
    paragraphs = [p.strip() for p in re.split(r'\n{2,}', text) if p.strip()]
    for para_idx, para in enumerate(paragraphs):
        chunk_id = f"{doc_base}_p{page_num}_para{para_idx+1:03d}"
        all_chunks.append({
            "chunk_id": chunk_id,
            "doc_base": doc_base,
            "page_num": page_num,
            "para_idx": para_idx+1,
            "text": para,
            "ocr_file": basename,
        })

df = pd.DataFrame(all_chunks)
df.to_csv("data/chunked_cleaned_text.csv", index=False)
```

```python
import re

def extract_heading(text):
    lines = str(text).split('\n')
    for line in lines[:2]:
        line_stripped = line.strip()
        if line_stripped.isupper() and len(line_stripped.split()) < 12:
            return line_stripped
        if re.match(r'^(SECTION|CLAUSE|CHAPTER|अनुभाग|धारा|PART)[\s\.\d\w]*', line_stripped, re.I):
            return line_stripped
    return ""

def label_chunk(text):
    if re.search(r"\b(alert|urgent|immediately|notice|attention|सावधान|सूचना|तत्काल)\b", text, re.I):
        return "alert"
    elif re.search(r"\b(shall|must|required to|responsible for|should|अवश्य|अनिवार्य|करना होगा|पाबंद|not allowed|prohibited|prohibit)\b", text, re.I):
        return "obligation"
    elif re.search(r"\b(regulation|rule|act|order|प्रावधान|विनियमन|अधिनियम|नियम)\b", text, re.I):
        return "regulation"
    else:
        return "other"

def postprocess_label(row):
    text = str(row.get("text", "")).lower().strip()
    orig_label = row.get("chunk_type", "").lower()

    alert_keywords = [
        "penalty", "contravention", "adjudication", "compounding",
        "violation", "fine", "breach", "enforcement", "liable", "subject to penalty"
    ]
    regulation_cues = [
        "shall have the same meaning as", "means", "definition", "explanation",
        "schedule", "eligible entity", "permitted", "prohibited", "limit", "tenor", "cost", "framework", "provided that"
    ]

    # Offload definitions and structural rules to REGULATION
    if any(cue in text for cue in regulation_cues):
        return "regulation"

    # Label prohibition/exception/override phrases as regulation unless there’s a penalty
    if re.search(r"\bshall(?:\s+not)?\s+be\s+deemed\b", text):
        return "regulation"

    # ALERT if penalty or contravention cues are present
    if any(k in text for k in alert_keywords):
        return "alert"

    # Still ALERT for strong prohibitions, even if original is obligation or regulation
    if orig_label in {"obligation", "regulation"} and (
        re.search(r"\bshall\s+not\b", text) or
        re.search(r"\bno person shall\b", text) or
        re.search(r"\bprohibited\b", text)
    ):
        return "alert"

    # Else, keep original label if valid, otherwise "other"
    return orig_label if orig_label in ["obligation", "regulation", "alert"] else "other"
```

```python
from keybert import KeyBERT
kw_model = KeyBERT()
def extract_keywords(text, top_n=8):
    if not isinstance(text, str) or len(text.strip()) == 0:
        return []
    keywords = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 3), stop_words='english', top_n=top_n)
    return [kw[0] for kw in keywords]
df['keywords'] = df['text'].apply(extract_keywords)
df.to_csv("data/chunked_with_keywords.csv", index=False)
```

```python
def context_rerank(query_idx, candidate_indices, orig_scores):
    q_row = df.iloc[query_idx]
    q_keywords = set(q_row['keywords']) if isinstance(q_row['keywords'], list) else set()
    q_section = q_row['section_path']
    q_type = q_row['post_label']  # CHANGED
    results = []
    for idx, base_score in zip(candidate_indices, orig_scores):
        cand = df.iloc[idx]
        bonus = 0
        c_keywords = set(cand['keywords']) if isinstance(cand['keywords'], list) else set()
        if q_keywords & c_keywords:
            bonus += 0.10
        if q_section and cand['section_path'] == q_section:
            bonus += 0.10
        # CHANGED both here to use post_label
        if (q_type, cand['post_label']) in [('obligation','regulation'), ('alert','regulation'),
                                            ('obligation','obligation'), ('alert','obligation')]:
            bonus += 0.05
        results.append((idx, base_score + bonus))
    results.sort(key=lambda x: x[1], reverse=True)
    reranked_indices = [t for t in results]
    reranked_scores = [t[1] for t in results]
    return reranked_indices, reranked_scores
```

```python
import pandas as pd
import networkx as nx
import json

# Load your annotations and links
df = pd.read_pickle("data/chunked_with_embeddings.pkl")
links_df = pd.read_csv("data/full_semantic_links.csv")

# Utility for chunk_id → doc_id
def get_doc_from_chunk(chunk_id):
    return chunk_id.split(".pdf")[0] + ".pdf"

# Add doc_id to nodes
df["doc_id"] = df["chunk_id"].map(get_doc_from_chunk)

# Export chunk nodes
node_metadata = {}
for idx, row in df.iterrows():
    node_metadata[row["chunk_id"]] = {
        "keywords": row.get("keywords", []),
        "summary": row.get("text", "")[:250],
        "section": row.get("section_path", ""),
        "type": row.get("chunk_type", ""),
        "language": row.get("language", ""),
        "doc_id": row.get("doc_id", "")
    }

# Add semantic and parent/child (hierarchy) edges
edges = []
for _, r in links_df.iterrows():
    edges.append({
        "source": r["source"],
        "target": r["target"],
        "similarity": r["score"],
        "link_type": r["link_type"]
    })

# (Optional) If you have section-parent-child relations, add them here:
# for idx, row in df.iterrows():
#     if row["parent_chunk_id"]:
#         edges.append({
#             "source": row["parent_chunk_id"],
#             "target": row["chunk_id"],
#             "link_type": "parent-child"
#         })

# Cluster/centrality
G = nx.Graph()
G.add_edges_from([(e["source"], e["target"]) for e in edges])
partitions = nx.algorithms.community.greedy_modularity_communities(G)
cluster_map = {}
for i, comm in enumerate(partitions):
    for node in comm:
        cluster_map[node] = i
deg_cent = nx.degree_centrality(G)

nodes = []
for n in G.nodes:
    meta = node_metadata.get(n, {})
    nodes.append({
        "id": n,
        "keywords": meta.get("keywords", []),
        "summary": meta.get("summary", ""),
        "section": meta.get("section", ""),
        "type": meta.get("type", ""),
        "cluster": cluster_map.get(n, 0),
        "centrality": deg_cent.get(n, 0),
        "label": n[:18] + "..." if len(n) > 21 else n,
        "language": meta.get("language", ""),
        "doc_id": meta.get("doc_id", "")
    })

# For document-level graph/network:
links_df["doc_source"] = links_df["source"].map(get_doc_from_chunk)
links_df["doc_target"] = links_df["target"].map(get_doc_from_chunk)

doc_links = (links_df.groupby(['doc_source', 'doc_target'])
    .agg({'score': 'mean','link_type': 'count'})
    .rename(columns={'link_type':'link_count','score':'avg_score'}).reset_index()
)
doc_edges = [
    {
        "source": row["doc_source"],
        "target": row["doc_target"],
        "weight": row["link_count"],
        "avg_score": row["avg_score"]
    } for _, row in doc_links.iterrows() if row["doc_source"] != row["doc_target"]
]
doc_nodes = [{"id": doc, "label": doc} for doc in sorted(set(links_df["doc_source"]) | set(links_df["doc_target"]))]
# Save to current directory
with open("semantic_graph_export.json", "w") as f:
    json.dump({"nodes": nodes, "edges": edges}, f, indent=2)
with open("document_graph_export.json", "w") as f:
    json.dump({"nodes": doc_nodes, "edges": doc_edges}, f, indent=2)

# Enable download to your local machine (for both files)
from google.colab import files
files.download('semantic_graph_export.json')
files.download('document_graph_export.json')
```
